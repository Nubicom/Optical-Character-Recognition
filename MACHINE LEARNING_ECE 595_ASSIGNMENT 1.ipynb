```python
# Use Tesseract to recognize characters in each image.
import cv2
import os
from PIL import Image
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files (x86)\Tesseract-OCR\tesseract'
rootdir = r"C:\Users\nubil\Desktop\ECE 59500 MACHINE LEARNING\pfw images"

for filename in os.listdir(rootdir):
    if not filename.endswith("webp"):
        full_image_path = os.path.join(rootdir,filename)
        img = cv2.imread(full_image_path)
        text = pytesseract.image_to_string(img)
        print(text)
```

```python
# Display images and recognized text.
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from PIL import Image
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files (x86)\Tesseract-OCR\tesseract'
rootdir = r"C:\Users\nubil\Desktop\ECE 59500 MACHINE LEARNING\pfw images"

for filename in os.listdir(rootdir):
    if not filename.endswith("webp"):
        full_image_path = os.path.join(rootdir,filename)
        img = cv2.imread(full_image_path)
        plt.show()
        plt.imshow(img)
        text = pytesseract.image_to_string(img)
        print(text)
```

```python
# Save the image file names and recognized text in a CSV file
import os
import csv
with open("result.csv", "w") as file:
    for filename in os.listdir(rootdir):
        if not filename.endswith("webp"):
            full_image_path = os.path.join(rootdir,filename)
            
            img = cv2.imread(full_image_path)
#           img = cv2.resize(img, (700,300))
#           img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #image enhancement messed up the output of the recognized text
            text = pytesseract.image_to_string(img)
            writer = csv.writer(file)
            data = [filename,text]
            writer.writerow(data)
```
